```python
data_used_gb = float(input("Enter total data used (GB): "))
price_per_gb = float(input("Enter price per GB: "))

total_bill = round(data_used_gb * price_per_gb, 2)
cost_per_mb = round(price_per_gb/1024, 4)
projected_usage_gb = round(data_used_gb * 1.2, 2)
projected_cost = round(projected_usage_gb * price_per_gb, 2)

print(f"Total Bill: $ {total_bill}")
print(f"Average Cost per MB: $ {cost_per_mb}")
print(f"Projected Cost: $ {projected_cost} for {projected_usage_gb} GB")
```

```python
password = input("Enter you password")
special_char = set('%@&#!^*()_-')
def has_upper(string):
    for char in string:
        if char.isupper():
            return True
    return False
def has_num(string):
    for char in string:
        if char.isdigit():
            return True
    return False
def check_password(password):
    errors = []
    if len(password)<8:
        errors.append("Length less than 8")
    if not has_num(password):
        errors.append("No Numbers")
    if  not has_upper(password):
        errors.append("No upper case latters")
    if not any( c in password for c in special_char):
        errors.append("No special characters")
    return errors

errors = check_password(password)

if not errors:
    print("strong password")
else:
    print(f"Weak Password : {errors}")
```

```python
nested_list = [ 1 , [2,3] , [4,[5,6]] , [7,[8,9]] ]

def flatten(nested_list):
    result = []
    for item in nested_list:
        if type(item) ==  list :
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

flattened_list = flatten(nested_list)
print(flattened_list)

for item in nested_list:
    print(type(item))
```

```python
def format_reasons(reasons):
    if not reasons:
        return ""
    if len(reasons) == 1:
        return reasons[0]
    else:
        return ', '.join(reasons[:-1]) + ' and ' + reasons[-1]

age  = int(input("Enter your age:"))
monthly_Income = float(input("Enter your monthly income:"))
credit_score = int(input("Enter your credit score :"))

def check_loan_eligibility(age, monthly_Income, credit_score):
    rejection_cause = []
    if age < 21 :
        rejection_cause.append("Age less than 21")
    if monthly_Income < 30000:
        rejection_cause.append("Income less than 30000")
    if credit_score < 650:
        rejection_cause.append("Credit score less than 650")

    formatted_reasons = format_reasons(rejection_cause)
    if not rejection_cause:
        print("Loan Approved")
    else:
        print(f"Loan Denied : {formatted_reasons}")

check_loan_eligibility(age, monthly_Income, credit_score)
```

```python
def GCD(a, b):
    while b:
        a, b = b, a % b
    return a

num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
gcd = GCD(num1, num2)
print(f"GCD of {num1} and {num2} is {gcd}")
```
